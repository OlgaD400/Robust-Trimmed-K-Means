```python
rtkm_f1_data = [np.average(f1_wbc_rtkm), np.average(f1_shuttle_rtkm), np.average(f1_yeastout_rtkm),
                np.average(f1_sceneout_rtkm), np.average(f1_emotions_rtkm), 
                np.average(f1_sceneout_normal_rtkm), np.average(f1_yeast_normal_rtkm), np.average(f1_emotions_normal_rtkm),
                .325, .598, .399]

kmor_f1_data = [np.average(f1_wbc_kmor), np.average(f1_shuttle_kmor), np.average(f1_yeastout_kmor), 
                np.average(f1_sceneout_kmor),  np.average(f1_emotions_kmor), 
                np.average(f1_sceneout_normal_kmor), np.average(f1_yeast_normal_kmor), np.average(f1_emotions_normal_kmor),
                .161, .598, .311]

neo_f1_data = [np.average(f1_wbc_neo), np.average(f1_shuttle_neo), np.average(f1_yeastout_neo), 
               np.average(f1_sceneout_neo), np.average(f1_emotions_neo), 
               np.average(f1_sceneout_normal_neo), np.average(f1_yeast_normal_neo), np.average(f1_emotions_normal_neo),  
               .366, .626, .373]


rtkm_me_data = [np.average(me_wbc_rtkm), np.average(me_shuttle_rtkm), np.average(me_yeastout_rtkm), 
                np.average(me_sceneout_rtkm), np.average(me_emotions_rtkm), 
               np.average(me_sceneout_normal_rtkm), np.average(me_yeast_normal_rtkm), np.average(me_emotions_normal_rtkm)]

kmor_me_data = [np.average(me_wbc_kmor), np.average(me_shuttle_kmor), np.average(me_yeastout_kmor), 
                np.average(me_sceneout_kmor), np.average(me_emotions_kmor),
               np.average(me_sceneout_normal_kmor), np.average(me_yeast_normal_kmor),np.average(me_emotions_normal_kmor)]

neo_me_data = [np.average(me_wbc_neo), np.average(me_shuttle_neo), np.average(me_yeastout_neo), 
               np.average(me_sceneout_neo),np.average(me_emotions_neo), 
              np.average(me_sceneout_normal_neo), np.average(me_yeast_normal_neo),  np.average(me_emotions_normal_neo), ]


f1_table_data = {'rtkm': rtkm_f1_data, 'kmor': kmor_f1_data, 'neo': neo_f1_data}
me_table_data = {'rtkm': rtkm_me_data, 'kmor': kmor_me_data, 'neo': neo_me_data}

f1 = pd.DataFrame(f1_table_data, index=['wbc', 'shuttle', 'yeast out', 'scene out',
                                       'emotions out', 'scene normal', 'yeast normal', 'emotions_normal', 
                                        'scene', 'yeast', 'emotions'])
me = pd.DataFrame(me_table_data, index=['wbc', 'shuttle', 'yeast out', 'scene out',
                                       'emotions', 'scene normal', 'yeast normal', 'emotions_normal'])

```

```python
me = me.round(2)
me
```

```python
me_ranked = me.rank(axis = 1)
me_ranked
```

```python
me_mat = me_ranked.to_numpy()
n = me_mat.shape[0]
print(n)
r_bar = np.sum(me_mat, axis = 0)/n
print(r_bar)

inner_sum = np.sum((r_bar - 2)**2)
    
Q_stat = n * inner_sum

print(Q_stat)
```

```python
# Nemenyi statistic 
CD = np.sqrt(12/(8*6))*2.343/np.sqrt(2)
print(CD<=(r_bar[2]-r_bar[0]), CD<=(r_bar[2]-r_bar[1]), \
      CD<=(r_bar[1]-r_bar[0]) )
```

```python
import scikit_posthocs as sp

data = np.array([me_ranked['rtkm'], me_ranked['kmor'], me_ranked['neo']])

#perform Nemenyi post-hoc test
sp.posthoc_nemenyi_friedman(data.T)
```
